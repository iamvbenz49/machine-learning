```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph
## plotting a graph using plt.plot(x axis,y axis)

```python
x = [1,2,3]
y = [4,5,6]
plt.title("First Graph",color="r")

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks(x) #x axis values
plt.yticks(y) #y axis values

plt.plot(x,y,label="1x",color='r',marker='2',linestyle='--') #for scake
plt.legend()
```

### why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0; therefore, the x data are [0, 1, 2, 3].

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
x2 = np.arange(0,20)

#resizing graph
plt.figure(figsize=(15,7),dpi=100)

plt.plot(x2,x2**2)
plt.savefig("graph.png",dpi=100)
```

# **Bar Chart**

```python
labels = ['A','B','C']
values = [1,5,3]


#plotting barcharts
bars = plt.bar(labels,values)

#setting hatches "*+-./OX\ox|
# bar[0].set_hatch("x")
# bar[1].set_hatch("/")
# bar[2].set_hatch("-")

patterns = ['/','-','*']
for bar in bars:
  bar.set_hatch(patterns.pop())

plt.show()
```
